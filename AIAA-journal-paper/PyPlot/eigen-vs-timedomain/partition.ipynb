```python
N=100
x=np.zeros(N)
y=np.zeros(N)
for i in np.arange(N):
    x[i]=np.cos(i/N*2*3.1415926*1.1);
    y[i]=np.sin(i/N*2*3.1415926*1.1);

    
outputfile='unstable-case.pdf'
    
fig = plt.figure(figsize=plt.figaspect(1))
plt.title('$\lambda = -1+j $');
plt.plot(x,y,'--k',label='unit circle')
plt.ylim(-1.5,1.5)
plt.xlim(-1.5,1.5)


a=-1;
b=1;
N=300;
lam_re=np.zeros(N);
lam_im=np.zeros(N);
for i in np.arange(N):
    t=0.1*i
    lam_re[i]=1-a*t;
    lam_im[i]=-b*t;    
    #denom=(1-a*t)*(1-a*t)+(b*t)*(b*t)
    #lam_re[i]=( 1-a*t)/denom;
    #lam_im[i]=(b*t)/denom;
#plt.plot(lam_re,lam_im,'r-o',label='$1-T\lambda$')



lam_re=np.zeros(N);
lam_im=np.zeros(N);
for i in np.arange(N):
    t=0.1*i
    #lam_re[i]=1-a*t;
    #lam_im[i]=-b*t;    
    denom=(1-a*t)*(1-a*t)+(b*t)*(b*t)
    lam_re[i]=( 1-a*t)/denom;
    lam_im[i]=(b*t)/denom;
plt.plot(lam_re,lam_im,'b-s',label='$\dfrac{1}{1-T\lambda}$')    
plt.legend()


plt.grid()
plt.xlabel('Real part')
plt.ylabel('Imaginery part')
plt.tight_layout()

plt.text(1.05, -0.2, r'$T=0$',fontdict={'size': 12, 'color': 'r'})
plt.text(-0.0, -0.2, r'$T=+\infty$',fontdict={'size': 12, 'color': 'r'})


#plt.show()
pdf=PdfPages(outputfile)
pdf.savefig()
plt.close()
pdf.close()   
    
    
    
outputfile='stable-case.pdf'
fig = plt.figure(figsize=plt.figaspect(1))
plt.title('$\lambda = 1+j $');
plt.plot(x,y,'--k',label='unit circle')
plt.ylim(-1.5,1.5)
plt.xlim(-1.5,1.5)    
a=-a;
b=1;
lam_re=np.zeros(N);
lam_im=np.zeros(N);
for i in np.arange(N):
    t=0.1*i
    lam_re[i]=1-a*t;
    lam_im[i]=-b*t;    
    #denom=(1-a*t)*(1-a*t)+(b*t)*(b*t)
    #lam_re[i]=( 1-a*t)/denom;
    #lam_im[i]=(b*t)/denom;
#plt.plot(lam_re,lam_im,'r-o',label='$1-T\lambda$')

 

lam_re=np.zeros(N);
lam_im=np.zeros(N);
for i in np.arange(N):
    t=0.1*i
    #lam_re[i]=1-a*t;
    #lam_im[i]=-b*t;    
    denom=(1-a*t)*(1-a*t)+(b*t)*(b*t)
    lam_re[i]=( 1-a*t)/denom;
    lam_im[i]=(b*t)/denom;
plt.plot(lam_re,lam_im,'b-s',label='$\dfrac{1}{1-T\lambda}$')    
#plt.scatter(lam_re,lam_im,c=range(len(lam_re)),marker='_', s=2,label='$\dfrac{1}{1-T\lambda}$')
plt.legend()
plt.grid()
plt.xlabel('Real part')
plt.ylabel('Imaginery part')    
plt.tight_layout()
#plt.annotate(s ='$T=0$', xy=(1.05,-0.2),xytext=(0,0), xycoords='data',textcoords='offset points', fontsize=12)
#plt.annotate(s ='$T=+\infty$', xy=(-0.1,-0.2),xytext=(0,0), xycoords='data',textcoords='offset points', fontdict={'size': 12, 'color': 'r'})
plt.text(1.05, -0.2, r'$T=0$',fontdict={'size': 12, 'color': 'r'})
plt.text(-0, -0.2, r'$T=+\infty$',fontdict={'size': 12, 'color': 'r'})

#plt.show()
pdf=PdfPages(outputfile)
pdf.savefig()
plt.close()
pdf.close()    


```

```python
p=np.loadtxt('unsteady-vs-steady-18d5kpa')
outputfile='annular-cascade-eff.pdf'

fig = plt.figure(figsize=plt.figaspect(0.2))
plt.plot(p[:,0],p[:,1],'r-o',label='unsteady')
plt.plot(p[:,0],p[:,3],'b-o',label='steady')
plt.legend()

fig = plt.figure(figsize=plt.figaspect(0.2))
plt.plot(p[:,0],p[:,3],'b-o',label='steady')
plt.legend()

fig = plt.figure(figsize=plt.figaspect(0.2))

freq = np.arange(p.shape[0]);
sp=np.fft.fft(p[:,1]-np.average(p[:,1]));
plt.plot(freq,np.absolute(sp[:]),'g-x',markersize=8,label='unsteady')
sp=np.fft.fft(p[:,3]-np.average(p[:,3]));
plt.plot(freq,np.absolute(sp[:]),'r-o',markersize=8,label='steady');
sp=np.fft.fft(p[:,5]-np.average(p[:,5]));
plt.plot(freq,np.absolute(sp[:]),'b-+',markersize=8,label='perturbation');
plt.yscale('log')
plt.grid()
plt.legend()

plt.xlim(1,100);
plt.ylim(1e3,1e7);
```
