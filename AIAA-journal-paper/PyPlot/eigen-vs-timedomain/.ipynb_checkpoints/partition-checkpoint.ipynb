```python
N=100
x=np.zeros(N)
y=np.zeros(N)
for i in np.arange(N):
    x[i]=np.cos(i/N*2*3.1415926*1.1);
    y[i]=np.sin(i/N*2*3.1415926*1.1);

    
outputfile='unstable-case.pdf'
    
fig = plt.figure(figsize=plt.figaspect(1))
plt.title('$\lambda = -1+j $');
plt.plot(x,y,'--k',label='unit circle')
plt.ylim(-1.5,1.5)
plt.xlim(-1.5,1.5)


a=-1;
b=1;
N=300;
lam_re=np.zeros(N);
lam_im=np.zeros(N);
for i in np.arange(N):
    t=0.1*i
    lam_re[i]=1-a*t;
    lam_im[i]=-b*t;    
    #denom=(1-a*t)*(1-a*t)+(b*t)*(b*t)
    #lam_re[i]=( 1-a*t)/denom;
    #lam_im[i]=(b*t)/denom;
#plt.plot(lam_re,lam_im,'r-o',label='$1-T\lambda$')



lam_re=np.zeros(N);
lam_im=np.zeros(N);
for i in np.arange(N):
    t=0.1*i
    #lam_re[i]=1-a*t;
    #lam_im[i]=-b*t;    
    denom=(1-a*t)*(1-a*t)+(b*t)*(b*t)
    lam_re[i]=( 1-a*t)/denom;
    lam_im[i]=(b*t)/denom;
plt.plot(lam_re,lam_im,'b-s',label='$\dfrac{1}{1-T\lambda}$')    
plt.legend()


plt.grid()
plt.xlabel('Real part')
plt.ylabel('Imaginery part')
plt.tight_layout()

plt.text(1.05, -0.2, r'$T=0$',fontdict={'size': 12, 'color': 'r'})
plt.text(-0.0, -0.2, r'$T=+\infty$',fontdict={'size': 12, 'color': 'r'})


#plt.show()
pdf=PdfPages(outputfile)
pdf.savefig()
plt.close()
pdf.close()   
    
    
    
outputfile='stable-case.pdf'
fig = plt.figure(figsize=plt.figaspect(1))
plt.title('$\lambda = 1+j $');
plt.plot(x,y,'--k',label='unit circle')
plt.ylim(-1.5,1.5)
plt.xlim(-1.5,1.5)    
a=-a;
b=1;
lam_re=np.zeros(N);
lam_im=np.zeros(N);
for i in np.arange(N):
    t=0.1*i
    lam_re[i]=1-a*t;
    lam_im[i]=-b*t;    
    #denom=(1-a*t)*(1-a*t)+(b*t)*(b*t)
    #lam_re[i]=( 1-a*t)/denom;
    #lam_im[i]=(b*t)/denom;
#plt.plot(lam_re,lam_im,'r-o',label='$1-T\lambda$')

 

lam_re=np.zeros(N);
lam_im=np.zeros(N);
for i in np.arange(N):
    t=0.1*i
    #lam_re[i]=1-a*t;
    #lam_im[i]=-b*t;    
    denom=(1-a*t)*(1-a*t)+(b*t)*(b*t)
    lam_re[i]=( 1-a*t)/denom;
    lam_im[i]=(b*t)/denom;
plt.plot(lam_re,lam_im,'b-s',label='$\dfrac{1}{1-T\lambda}$')    
#plt.scatter(lam_re,lam_im,c=range(len(lam_re)),marker='_', s=2,label='$\dfrac{1}{1-T\lambda}$')
plt.legend()
plt.grid()
plt.xlabel('Real part')
plt.ylabel('Imaginery part')    
plt.tight_layout()
#plt.annotate(s ='$T=0$', xy=(1.05,-0.2),xytext=(0,0), xycoords='data',textcoords='offset points', fontsize=12)
#plt.annotate(s ='$T=+\infty$', xy=(-0.1,-0.2),xytext=(0,0), xycoords='data',textcoords='offset points', fontdict={'size': 12, 'color': 'r'})
plt.text(1.05, -0.2, r'$T=0$',fontdict={'size': 12, 'color': 'r'})
plt.text(-0, -0.2, r'$T=+\infty$',fontdict={'size': 12, 'color': 'r'})

#plt.show()
pdf=PdfPages(outputfile)
pdf.savefig()
plt.close()
pdf.close()    


```

```python
import math
p2=np.loadtxt('sensor1_16d7.dat')

fig = plt.figure(figsize=plt.figaspect(1))

#print(2000/1680)
p_fit=np.zeros(np.size(p2[:,2]));
omega=2000;
omega=1680*1.1835;
phi=np.pi/1.54;
c0=7.0e-12;
lam=33.936;
#print(1680*0.0202)
for n in range(np.size(p_fit)):
    t=p2[n,0];
    p_fit[n]=c0*math.exp(t*lam)*np.cos(omega*t+phi);
            
plt.plot(p2[:,0],np.log10(np.abs(p_fit)),'k-',markersize=1,label='16.7kPa fitted');
plt.plot(p2[:,0],np.log10(np.abs(p2[:,2]-p2[0,2])+1e-11),'r.',markersize=1,label='16.7kPa');
#plt.plot(p2[:,0],p_fit,'k-',label='16.7kPa fitted');
#plt.plot(p2[:,0],p2[:,2]-p2[0,2],'r',label='16.7kPa');

e=p2[:,2]-p_fit;
print(np.sum(e)*1e10)

plt.grid();plt.legend()
plt.xlabel('Time');plt.ylabel('Unsteady pressure')
plt.xlim(0.,.8);
plt.ylim(-14,0);
plt.tight_layout();
plt.show()
#pdf=PdfPages(outputfile)
#pdf.savefig()
#plt.close()
#pdf.close()

fig = plt.figure(figsize=plt.figaspect(1))
plt.plot(p2[:,0],p_fit-(p2[:,2]-p2[0,2]),'r.',markersize=1,label='16.7kPa delta');
plt.plot(p2[:,0],p_fit,'k.',markersize=1,label='16.7kPa fitted');
plt.grid();plt.legend()
plt.xlim(0.6,.8);
plt.tight_layout();
plt.show()
```

```python
import math
p1=np.loadtxt('sensor1_16d6.dat')

fig = plt.figure(figsize=plt.figaspect(1))

# lambda = (0.0096, 1.1863)
p_fit=np.zeros(np.size(p1[:,2]));
omega=1680*1.1863;
phi=np.pi*-0.2;
c0=3e-15;
lam=1680*0.0096;
for n in range(np.size(p_fit)):
    t=p1[n,0];
    p_fit[n]=c0*math.exp(t*lam)*np.cos(omega*t+phi);

plt.plot(p1[:,0],np.log10(np.abs(p_fit)),'k-',markersize=1,label='16.6kPa fitted');
plt.plot(p1[:,0],np.log10(np.abs(p1[:,2]-p1[0,2])+1e-10),'r.',markersize=1,label='16.6kPa');
#plt.plot(p1[:,0],p1[:,2]-p1[0,2],'r',label='16.6kPa');
#plt.plot(p1[:,0],p_fit,'k-',label='16.6kPa fitted');

e=p1[:,2]-p_fit;
print(np.sum(e)*1e10)

plt.grid();plt.legend()
plt.xlabel('Time');plt.ylabel('Unsteady pressure')
plt.xlim(0.,2);
plt.ylim(-16,0);
plt.tight_layout();plt.show()
#pdf=PdfPages(outputfile)
#pdf.savefig()
#plt.close()
#pdf.close()  
```
